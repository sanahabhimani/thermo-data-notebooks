```python
# is there more x001 data? look in notebook
lsroxtemps = all_verify_data[0]['CAL']['temp_avg']
x001_1 = all_verify_data[0]['X-001']['temp_avg']
x001_2 = all_verify_data[2]['X-001']['temp_avg']
x001_3 = all_verify_data[3]['X-001']['temp_avg'] # has 51.7mK and 60mK; no 55
```

```python
%matplotlib inline
matplotlib.rcParams['figure.figsize']=[10, 8]
```

```python
#not including 3rd batch of x001 data
x001 = np.concatenate((x001_1, x001_2[0:30])) 
lsrox = np.concatenate((lsroxtemps, lsroxtemps))
```

```python
lsrox_sort = lsrox[np.argsort(x001)]
x001_sort = x001[np.argsort(x001)]
```

```python
x = np.arange(0,10)
plt.plot(lsrox_sort, x001_sort, '.')
plt.plot(x, label = 'one-to-one line')
plt.xlabel("LS ROX Temperatures [K]")
plt.ylabel("X001 Temperatures [K]")
plt.ylim(0.04,0.205)
plt.xlim(0.04,0.205) #not the linear fit; just a raw one to one line
plt.legend()
```

# Linear Fit

```python
linfit = np.polyfit(lsrox_sort, x001_sort, 1)
print(linfit)
```

```python
p = np.poly1d(linfit)
p_eval = p(np.linspace(np.min(lsrox_sort), np.max(lsrox_sort), len(lsrox_sort)))
print(len(p_eval))
print(p)
```

```python
plt.plot(np.linspace(np.min(lsrox_sort), np.max(lsrox_sort), len(lsrox_sort)), p_eval, color = 'r', label = 'Linear Fit')
plt.plot(lsrox_sort, x001_sort, '.')
plt.ylim(0.04,0.205)
plt.xlim(0.04,0.205) 
plt.legend(fontsize = 14)
plt.xlabel("LSCALROX Temperatures [K]", fontsize = 14)
plt.ylabel("X001 Temperatures [K]", fontsize = 14)
plt.title("X001 Verification Temperature Run", fontsize = 14)
```

### Calculate standard deviations

```python
x001_1 = np.sort(x001_1)
x001_2 = np.sort(x001_2[0:30])
x001_array = {'x001_1': x001_1, 
              'x001_2': x001_2}
```

```python
p_eval_2[4]
```

```python
resids_55mk = []
for i in x001_array:
    print ('a check: ', x001_array[i][0])
    print ('linfit value', p_eval_2[0])
    resids = x001_array[i][0] - p_eval_2[0]
    resids_55mk.append(resids)
    
print(resids_55mk)
print('stand dev: ', np.std(resids_55mk))
```
